This is a seven-segment detector that is able to detect the digit from its seven-segment image

```python
import cv2
import matplotlib.pyplot as plt
cut = cv2.imread("/content/2.jpg");
original = cv2.cvtColor(cut, cv2.COLOR_BGR2GRAY)
plt.imshow(original);

h, w= original.shape
print(h, w)
```

Breaking the image into seven parts corresponding to seven segments and the detecting the dominant pixel value after quantisation; thereby finding if it is ON or OFF

```python
import numpy as np
hor1x1=0
hor1x2=-1
hor1y1=0
hor1y2=h//10
cut=original.copy()
seg1=cut[hor1y1:hor1y2, hor1x1:hor1x2]

for i in range(len(seg1)):
  for j in range(len(seg1[i])):
    if seg1[i][j]>220:
      seg1[i][j]=0
    else:
      seg1[i][j]=255
seg1=np.array(seg1)
plt.imshow(seg1);

seg1=np.array(seg1).flatten()
seg1=list(seg1)

h1=False
a = seg1.count(0)
b = seg1.count(255)
if a>b:
  h1=False
else:
  h1=True
print(h1)
```
